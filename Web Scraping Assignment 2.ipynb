```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings 
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.shine.com/")
```

```python
designation=driver.find_element(By.CLASS_NAME,"form-control  ")


designation.send_keys('Data Analyst')
```

```python
location=driver.find_element(By.ID,"id_loc")
location.send_keys('Banglore')
```

```python
Exp=driver.find_element(By.ID,"id_exp")
Exp.send_keys('')
```

```python
Search=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button")
Search.click()
```

```python
link=[]
anchor_elements =driver.find_elements(By.TAG_NAME, "a")
for element in anchor_elements:
    href = element.get_attribute('href')
    if href is not None:
        if 'jobs' and 'data' and 'analyst' in href:
            if 'job-search' not in href:
                link.append(href)
```

```python
link
```

```python
New_link=set(link)
```

```python
job_titles=[]
job_location=[]
campany_name=[]
experience_required=[]
```

```python
for i in New_link:
    driver.get(i)
    
    try:
        a=driver.find_element(By.CLASS_NAME,'font-size-24').text
        job_titles.append(a)
    except:
        job_titles.append('NA')
        
    try:
        b=driver.find_element(By.CLASS_NAME,'JobDetailWidget_jobCard_cName__qvsdW').text
        campany_name.append(b)
    except:
        campany_name.append('NA')
    
    try:
        c=driver.find_element(By.XPATH,'/html/body/div/div[1]/div/div/div[1]/div[1]/div[2]/div/div[3]/div[2]').text
        experience_required.append(c)
    except:
        experience_required.append('NA')
        
    try:
        d=driver.find_element(By.XPATH,'/html/body/div/div[1]/div/div/div[1]/div[1]/div[2]/div/div[3]/div[1]/a').text
        job_location.append(d)
    except:
        job_location.append('NA')
        
```

```python
job_location
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({"job_titles":job_titles,"job_location":job_location,"company_name":campany_name,"experience":experience_required})
```
